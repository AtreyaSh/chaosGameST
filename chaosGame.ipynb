# Chaos game with an equilateral triangle

## Background

Chaos games are essentially random games with a fixed set of instructions. In the case of this example, we will see how chaos games can produce beautiful fractal structures. More information on chaos games can be found here:

https://en.wikipedia.org/wiki/Sierpinski_triangle#Chaos_game

The motivation for this project was this brilliant [video](https://www.youtube.com/watch?v=kbKtFN71Lfs&t=77s) produced by Numberphile. I would recommend watching this for a better context in this project.

## Algorithm

1. Draw an equilateral triangle without its edges. In this case, just plot 3 points for the 3 vertices. Let these be called `A`, `B` and `C` respectively. An example of this is shown in the figure below.

2. Define a random point within the equilateral triangle. 

3. Randomly choose one of the vertices `A`, `B` or `C`.

4. Draw another point halfway between the defined random point from step 2 and the randomly chosen vertex.

5. Repeat this process for `N` iterations.

This iterative process should lead to a beautiful convergence. Let's try this out below!

```python
# First we need to define some environmental variables

# stores plots within notebook
%matplotlib inline 

# importing necessary packages
import random
import numpy as np
import time
import matplotlib.pyplot as plt
import os
from IPython import display

# set default figure size
plt.rcParams['figure.figsize'] = (10.0, 8.0)

# crate the basic plot
plt.scatter([0,2,4], [0,4,0])
bounds = ([0,0],[2,4],[4,0])
plt.xlim([-0.5,4.5])
plt.ylim([-0.5,4.5])
```

## Chaos game function

Here we define our chaos game function. This process essentially cotinuously produces new scatter plots and displays the latest one. The end result is a kind of quasi-animation. 

With all honestly, this code is not entirely efficient, but it nonetheless serves the purpose of showing what happens in this chaos game. Increased efficiency for this code is the next priority.

```python
def Chaos_Game(start=None,n=None):
    
    plt.scatter([0,2,4], [0,4,0])
    bounds = ([0,0],[2,4],[4,0])
    plt.xlim([-0.5,4.5])
    plt.ylim([-0.5,4.5])
    
    if start==None:
        # default starting point
        start=np.array([2,2])
    if n==None:
        # default number of iterations
        n=500
        
    for i in xrange(1,n+1):
        plt.scatter(start[0],start[1],lw='0',s=19)
        plt.draw()

        if i != n:
            print(i)
            display.clear_output(wait=True)
            display.display(plt.gcf())
        else:
            print(i)
            display.clear_output(wait=True)

        rnc=random.choice(bounds)
        rnc=np.array(rnc)
        start=start+(0.5*(rnc-start))
    
    # save the final figure
    plt.savefig(os.getcwd()+"/images/chaos_game"+str(n)+".png")
    return None
```
